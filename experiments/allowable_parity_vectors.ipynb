```python
import sys, os
sys.path.insert(0, os.path.abspath('../'))
from coreli import *
import itertools 
from functools import reduce
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_values = []
y_values = []

# Define the range for x and y
for x in range(0, 101):  # x >= 0
    for y in range(0, 101):  # y >= 0
        if 2**(x + y) > 3**y and y >= x:
            x_values.append(x)
            y_values.append(y)

# Scatter plot
plt.scatter(x_values, y_values, color='blue', label='Points satisfying conditions')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of points satisfying 2^(x+y) > 3^y and y >= x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the size of the pixel map
width = max(x_values) + 1
height = max(y_values) + 1

# Create a 2D numpy array for the pixel map
pixel_map = np.zeros((height, width), dtype=int)

# Set pixels to black (1) if the conditions are satisfied
for x, y in zip(x_values, y_values):
    pixel_map[y, x] = 1

# Display the pixel map
plt.imshow(pixel_map, cmap='gray', origin='lower')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Black and White Pixel Map')
plt.colorbar(label='Pixel Value')
plt.show()
```
